```python
import os
import numpy as np
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import load_model
```

```python
def img_to_nparray(file_path, data_dir) :
    return(img_to_array(load_img(file_path, target_size = (200,200))))
```

```python
data_dir = "./test_set/"
image_files = [data_dir+file_name for file_name in os.listdir(data_dir) if not file_name.startswith('.')]
images = [img_to_nparray(image, data_dir)/255 for image in image_files]
X_test = np.asarray(images)
```

```python
model = load_model('model_01102019.h5')
```

```python
Y_test = model.predict(X_test)
nb_images = len(Y_test)
```

```python
from PIL import Image
import time

marque = ["Barilla", "Bjorg", "Gerble", "Panzani"]

for i in range(nb_images):
    
    image = (X_test[i,:,:,:]*255).astype('uint8')
    image = Image.fromarray(image)
    display(image)
    
    print(marque[Y_test.tolist()[i].index(max(Y_test[i]))])
    
    time.sleep(2)
```
